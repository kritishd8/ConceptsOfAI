```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def df(x):
    return 2 * x

def gradient_descent(starting_point, learning_rate, iterations):
    x = starting_point
    steps = [x]
    
    for _ in range(iterations):
        grad = df(x)
        x = x - learning_rate * grad
        steps.append(x)
    
    return steps


starting_point = 5.0
learning_rate = 0.1
iterations = 50

steps = gradient_descent(starting_point, learning_rate, iterations)

x_values = np.linspace(-6, 6, 400)
y_values = f(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label="f(x) = x^2", color="black")
plt.scatter(steps, f(np.array(steps)), color="blue", label="Steps of Gradient Descent")
plt.plot(steps, f(np.array(steps)), linestyle="--", color="blue")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent on f(x) = x^2")
plt.legend()
plt.grid()
plt.show()
```
